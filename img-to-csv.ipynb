```python
#get image from outter folder and save it in the csv

from PIL import Image ,ImageOps
import numpy as np
import sys
import os
import csv

# default format can be changed as needed
def createFileList(myDir, format='.png'):
    fileList = []
    print(myDir)
    for root, dirs, files in os.walk(myDir, topdown=False):
        for name in files:
            if name.endswith(format):
                fullName = os.path.join(root, name)
                fileList.append(fullName)
    return fileList

# load the original image
myFileList = createFileList('./smallpic/test/1') 

for file in myFileList:
    print(file)
    img_file = Image.open(file)
    # img_file.show()

    # get original image parameters...
    width, height = img_file.size
    format = img_file.format
    mode = img_file.mode

    # Make image Greyscale
    img_grey = ImageOps.grayscale(img_file)
    #img_grey.save('result.png')
    #img_grey.show()

    # Save Greyscale values
    #5120 * 221
    value = np.asarray(img_grey.getdata(), int).reshape((img_grey.size[1], img_grey.size[0]))
    value = value.flatten()
    test=list(value)
    test.append(1)
    with open("KidneyDataset.csv", 'a',newline='') as f:
        writer = csv.writer(f)
        writer.writerow(test)
    
    
```
